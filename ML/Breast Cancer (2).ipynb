```python
import numpy as np 
import pandas as pd 
import seaborn as sns
```

```python
data = pd.read_csv("data.csv")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
# clean data
data.drop(['Unnamed: 32', "id"], axis=1, inplace=True)
```

```python
data.diagnosis =[1 if value == "M" else 0 for value in data.diagnosis]
```

```python
data['diagnosis'] = data['diagnosis'].astype('category',copy=False)
plot = data['diagnosis'].value_counts().plot(kind='bar', title="Class distributions \n(0: Benign | 1: Malignant)")
fig = plot.get_figure()
```

```python
y = data["diagnosis"]
X = data.drop(["diagnosis"], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.30, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

```python
from sklearn.svm import SVC
svc=SVC() 
svc.fit(X_train,y_train)
y_pred2=svc.predict(X_test)
```

```python
accuracy2 = accuracy_score(y_test, y_pred2)
print(f'Accuracy: {accuracy2:.2f}')
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
y_pred3=dt.predict(X_test)
```

```python
accuracy3 = accuracy_score(y_test, y_pred3)
print(f'Accuracy: {accuracy3:.2f}')
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=0)
rfc.fit(X_train, y_train)
y_pred4 = rfc.predict(X_test)
```

```python
accuracy4 = accuracy_score(y_test, y_pred4)
print(f'Accuracy: {accuracy4:.2f}')
```

```python
from sklearn.ensemble import VotingClassifier
model = VotingClassifier(estimators=[('svc', svc), ('dt', dt),('rfc',rfc)], voting='hard')
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred4)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt=".0f", cmap="Blues")
plt.title("Confusion Matrix - Voting Classifier")
plt.xlabel("Predicted Class")
plt.ylabel("True Class")
plt.show()
```
